# Visualizing Data


### Import Library

```python
import matplotlib.pyplot as  plt
from collections import Counter
```

### Line Chart

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1057.9, 2862.5, 5979.6, 10289.7, 14958.3]

# create a line chat, years on x-axis and gdp on y-axis
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')

plt.title('Norminal GDP')
plt.ylabel('Billions of $')
plt.show()
```

### Bar Charts

```python
movies = ['Annie Hall', 'Ben-Hur', 'Casablanca', 'Gandhi', 'West Side Story']
num_oscar = [5, 11, 3, 8, 10]

# bars are shifted by 0.1, to make each bar centered
# default width of the bars  is 0.8, so we add 0.1 to the left coordinates
xs = [x + 0.1 for x, _ in enumerate(movies)]

# plot bars with left x-coordinate [xs], heights [num_oscar]
plt.bar(xs, num_oscar)

plt.ylabel('# of Academy Awards')
plt.title('My Favorite Movies')

plt.xticks([i + 0.1 for i, _ in enumerate(movies)], movies)

plt.show()
```

```python
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
decile = lambda grade: grade // 10 * 10
histogram = Counter(decile(grade) for grade in grades)

plt.bar([x - 4 for x in histogram.keys()],    # shift the bars to left by 4
        histogram.values(),                   # assign the corresponding values to the keys
        8)                                    # give each bar a width value of 8

plt.axis([-5, 105, 0, 5])                     # x-axis from -5 to 105 and y-axis from 0-5

plt.xticks([10 * i for i in range(11)])
plt.xlabel('Decile')
plt.ylabel('# of Students')
plt.title('Distribution of Exam 1 Grades')
plt.show()
```

```python
mentions = [500, 505]
years = [2013, 2014]

plt.bar([2012.6, 2013.6], mentions, 0.8)
plt.xticks([year-0.4   for year in years], years)
plt.ylabel('# of times I heard someone say \'data science\'')

#
#plt.ticklabel_format(useOffset=False)
plt.axis([2012.0, 2014.5, 499, 506])
plt.title('Look at the \'Huge\' Increase!')
plt.show()



```

```python
# the same chart without misleading y-axis
mentions = [500, 505]
years = [2013, 2014]

plt.bar([2012.6, 2013.6], mentions, 0.8)
plt.xticks([year-0.4   for year in years], years)
plt.ylabel('# of times I heard someone say \'data science\'')

plt.axis([2012.0, 2014.5, 0, 550])
plt.title('Not so Huge Anymore')
plt.show()
```

### Line Charts

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = variance[::-1]
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

# multiple plt.plot
plt.plot(xs, variance, 'g-', label='variance')
plt.plot(xs, bias_squared, 'r-.', label='bias^2')
plt.plot(xs, total_error, 'b:', label='total error')

# legend is automatically assigned because of the labels assigned
plt.legend(loc=9)
plt.xlabel('model comlexity')
plt.title('The Bias-Variance Tradeoff')
plt.show()
```

### Scatterplots

```python
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)

# label each point
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
        xy = (friend_count, minute_count),
        xytext=(5, -5),
        textcoords='offset points')
    
plt.title('Daily Minutes vs Number of Friends')
plt.xlabel('# of friends')
plt.ylabel('daily minutes spent on the site')
plt.show()
```
